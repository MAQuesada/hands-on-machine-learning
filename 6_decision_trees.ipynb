# **Decision Trees**

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data[:, 2:]  # petal length and width
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)

# show the tree
plt.figure(figsize=(8, 6))
plot_tree(tree_clf, feature_names=iris.feature_names, class_names=iris.target_names, fontsize=10)
plt.show()
```

```python
"""You can visualize the trained Decision Tree using the export_graphviz() method
to output a graph definition file called iris_tree.dot"""

from sklearn.tree import export_graphviz
from graphviz import Source


export_graphviz(
    tree_clf,
    out_file="iris_tree.dot",
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True)

Source.from_file("iris_tree.dot")
```

```python
"""Importance of features"""
import pandas as pd
pd.DataFrame({'predictors': iris.feature_names[:2], 'importance': tree_clf.feature_importances_})
```
